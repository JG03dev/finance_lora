```rust
extern crate reqwest;
extern crate serde;
extern crate serde_json;
extern crate tch;

use std::fs::File;
use std::io::{BufRead, BufReader};
use tch::{nn, nn::OptimizerConfig, Device, Tensor, kind};

fn load_and_preprocess_data(path: &str) -> Vec<(Tensor, Tensor)> {
    let file = File::open(path).expect("Unable to open file");
    let reader = BufReader::new(file);
    let mut data = Vec::new();

    for line in reader.lines() {
        let line = line.expect("Unable to read line");
        let split: Vec<&str> = line.split(',').collect();
        let text = split[0];
        let label = split[1].parse::<i64>().expect("Unable to parse label");

        let input_tensor = preprocess_text(text);
        let label_tensor = Tensor::of_slice(&[label]).to_device(Device::Cpu);

        data.push((input_tensor, label_tensor));
    }

    data
}

fn preprocess_text(text: &str) -> Tensor {
    Tensor::of_slice(&[0]) 
}

// fn create_lora_matrices(vs: &nn::VarStore, ranks: i64) -> Vec<Tensor> {
//     vec![]
// }

// fn apply_lora(prediction: Tensor, lora_matrices: &Vec<Tensor>) -> Tensor {asdkfaksdjfañlksd
//     prediction
// }

fn fine_tune_model(data: Vec<(Tensor, Tensor)>) {
    let vs = nn::VarStore::new(Device::Cpu);
    let gpt_model = tch::models::gpt2::GPT2LMHeadModel::new(&vs);

    let lora_ranks = 8;
    let lora_matrices = create_lora_matrices(&vs, lora_ranks);

    let mut opt = nn::Adam::default().build(&vs, 1e-4).unwrap();
    for epoch in 1..10 {
        for batch in data.iter() {
            let (input, target) = batch;
            let prediction = gpt_model.forward_t(input, true);

            let adapted_prediction = apply_lora(prediction, &lora_matrices);

            let loss = adapted_prediction.cross_entropy_for_logits(&target);
            opt.backward_step(&loss);
        }
        println!("Epoch: {}, Loss: {:?}", epoch, loss);
    }
}

fn evaluate_model(model: &tch::models::gpt2::GPT2LMHeadModel, test_data: &Vec<(Tensor, Tensor)>) {
    let mut total_loss = 0.0;
    for batch in test_data.iter() {
        let (input, target) = batch;
        let prediction = model.forward_t(input, false);
        let loss = prediction.cross_entropy_for_logits(&target);
        total_loss += f64::from(loss);
    }
    let avg_loss = total_loss / test_data.len() as f64;
    println!("Average Loss: {}", avg_loss);
}


let data = load_and_preprocess_data("invoices.csv");
fine_tune_model(data);

let test_data = load_and_preprocess_data("test_invoices.csv");
evaluate_model(&gpt_model, &test_data);

```
